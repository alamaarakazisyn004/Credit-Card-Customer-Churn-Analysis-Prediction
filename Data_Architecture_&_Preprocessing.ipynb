```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("BankChurners.csv")
df.head()
```

```python
# Understand the Data
df.info()
df.describe()
```

```python
# -------------------------------
# 1. Remove irrelevant columns
# -------------------------------
df = df.drop(columns=['CLIENTNUM'])

# Remove leakage columns
df = df.loc[:, ~df.columns.str.contains('Naive_Bayes')]
```

```python
# -------------------------------
# 2. Handle Missing Values
# -------------------------------
# Numerical → median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.median()))

# Categorical → mode
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# -------------------------------
# 3. Remove Duplicates
# -------------------------------
df = df.drop_duplicates()
```

```python
# -------------------------------
# 4. Handle Outliers (IQR Method)
# -------------------------------
def iqr_filter(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return data[(data[column] >= lower) & (data[column] <= upper)]

for col in ['Customer_Age', 'Credit_Limit', 'Total_Trans_Amt']:
    df = iqr_filter(df, col)
```

```python
# -------------------------------
# 5. Encode Target Variable
# -------------------------------
df['Attrition_Flag'] = df['Attrition_Flag'].map({
    'Existing Customer': 0,
    'Attrited Customer': 1
})
```

```python
# -------------------------------
# 6. One-Hot Encoding
# -------------------------------
df = pd.get_dummies(df, drop_first=True)
```

```python
# -------------------------------
# 7. Feature Scaling
# -------------------------------
scaler = StandardScaler()

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols = num_cols.drop('Attrition_Flag')

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# -------------------------------
# 8. Save Clean Dataset
# -------------------------------
df.to_csv("cleaned_dataset.csv", index=False)

print("Preprocessing Completed Successfully!")
```
